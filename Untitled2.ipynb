```python
#installing dependencies
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.style.use("seaborn-v0_8")
```

```python
symbol = input("Enter stock symbol (e.g., INFY.NS, RELIANCE.NS, ^NSEI): ").strip()

if not symbol:
    symbol = "INFY.NS"   # default if user presses enter

symbol
```

```python
df = yf.download(symbol, period="1y")

if df.empty:
    raise ValueError("No data found. Try another symbol like INFY.NS or TCS.NS.")

df = df[["Close"]].rename(columns={"Close": "close"})
df.head()
```

```python
# Daily returns
df["return"] = df["close"].pct_change()

# Annualized volatility (std * sqrt(252 trading days))
volatility = df["return"].std() * np.sqrt(252)

volatility
```

```python
plt.figure(figsize=(12,5))
plt.plot(df.index, df["close"], label="Closing Price", linewidth=2)
plt.title(f"{symbol} - 1-Year Price Chart")
plt.xlabel("Date")
plt.ylabel("Price")
plt.grid(True)
plt.show()
```

```python
df["rolling_vol"] = df["return"].rolling(20).std() * np.sqrt(252)

plt.figure(figsize=(12,5))
plt.plot(df.index, df["rolling_vol"], label="20-Day Rolling Volatility", color="purple", linewidth=2)
plt.title(f"{symbol} - Rolling Volatility (20-Day)")
plt.xlabel("Date")
plt.ylabel("Volatility")
plt.grid(True)
plt.show()
```

```python
print("===== VOLATILITY SUMMARY =====")
print(f"Stock: {symbol}")
print(f"Annualized Volatility: {volatility:.2%}")
print("\nInterpretation:")

if volatility > 0.40:
    print(" Very high volatility  price fluctuates significantly (high risk).")
elif volatility > 0.25:
    print(" Moderately high volatility  notable movement (medium-high risk).")
elif volatility > 0.15:
    print(" Normal volatility  typical for large-cap stocks.")
else:
    print(" Low volatility  stable stock with limited price movement.")
```
